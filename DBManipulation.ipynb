##Update Command for lat y long

```python
#pip install mysql-connector-python
import mysql.connector
import requests
import urllib.parse
import pandas as pd
```

```python
#conexion a base de datos 
conexion = mysql.connector.connect(user='XXX', password='XXX',
                              host='XXX',
                              database='XXX')
```

```python
#Traer data de oficinas (Solo código y direccion)
df = pd.read_csv('/content/drive/MyDrive/oficinas2.csv')
df
```

```python
#Almacenar lat y lon de cada direccion
lista_longitud = []
lista_latitud = []
```

```python
contador = 0
for x in df['Direccion'].values:
  print(x)
  
  try:

    url = 'https://nominatim.openstreetmap.org/search/' + urllib.parse.quote(x) +'?format=json'
    response = requests.get(url).json()
    latitud = response[0]["lat"]
    longitud = response[0]["lon"]

    lista_longitud.append(longitud)
    lista_latitud.append(latitud)


  except IndexError:
    lista_longitud.append('NULL')
    lista_latitud.append('NULL')


  
```

```python
print(lista_latitud)
print(lista_longitud)
```

```python
#Pasar las listas a cada columna
df['Latitud'] = lista_latitud
df['Longitud'] = lista_longitud 
```

```python
#Almacenar las sentencias de UPDATE para agregar la data a la tabla oficina
QUERY_COORDENADAS = []
```

```python
for i in range(df.shape[0]):
  QUERY_COORDENADAS.append('UPDATE X SET Latitud ="'+ df.iloc[i,2] + '", Longitud ="' + df.iloc[i,3] + '" WHERE CodOficina ='+ str(df.iloc[i, 0]))

  
```

```python
QUERY_COORDENADAS
```

```python
#Actualizar en sql workbench
mycursor = conexion.cursor()
for up_oficina in QUERY_COORDENADAS:
    mycursor.execute(up_oficina)
    conexion.commit()
```
